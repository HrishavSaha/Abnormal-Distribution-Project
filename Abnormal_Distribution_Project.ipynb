```python
import random
import plotly.express as plot
import plotly.figure_factory as ff
import plotly.graph_objects as go
import csv
import pandas as pd
import statistics
import termcolor
```

```python
def sampler(population_data, iterations):
  temp_population_data = []
  temp_population_data = population_data.tolist()
  sample_data = []
  mean = 0

  for i in range(0, iterations):
    rand = random.randint(0, len(temp_population_data))
    sample_data.append(temp_population_data[rand])
    temp_population_data.pop(rand)

  mean = statistics.mean(sample_data)

  return mean

def setup(iterations, population_data, sampler_iterations):
  mean_list = []

  for i in range(0, iterations):
    mean_list.append(sampler(population_data, sampler_iterations))
  
  return mean_list
```

```python
df = pd.read_csv("medium_data.csv")
data = df['reading_time']
df.head(10)
```

```python
population_mean = statistics.mean(data)
print("Population Mean: ", str(population_mean))
```

```python
sample_mean_list = setup(100, data, 30)
mean_mean = statistics.mean(sample_mean_list)
```

```python
plotter = ff.create_distplot([sample_mean_list], ["Mean"], show_hist=False)
plotter.add_trace(go.Scatter(x=[mean_mean, mean_mean], y=[0,1], mode='lines', name = 'Mean of Sample Means'))
plotter.add_trace(go.Scatter(x=[population_mean, population_mean], y=[0,1], mode='lines', name = 'Mean of Population Data'))
plotter.show()
```
